- FILTER THE DATA

```python
# The records where max_power is less than 1000

ucr[ucr['max_power(bhp)'] < 1000]
```

- GROUPBY

```python
# Group the DataFrame by the 'manufacturing_year' column and then get a specific group corresponding to the year 2020
ucr.groupby('manufacturing_year').get_group('2020')
```

- PIE CHART

```python
# Pie chart of column 'seats' grouped by the 'fuel_type' column
ucr.groupby(['fuel_type']).sum().plot( kind='pie', y='seats',autopct='%1.0f%%')
plt.show()
```

- DIST PLOT

```python
# Dist plot of column 'mileage(kmpl)'
plt.figure(figsize=(6,3.2),dpi=150)
plt.rc('xtick', labelsize=6)
plt.rc('ytick', labelsize=8)
sns.distplot(ucr['mileage(kmpl)'])
plt.show()
```

- LINE PLOT

```python
plt.plot(ucr['manufacturing_year'], ucr['engine(cc)'])
plt.xlabel('Manufacturing Year')
plt.ylabel('Engine (cc)')
plt.show()
```

- HIST PLOT

```python
# Hist plot of column 'fuel_type'
sns.histplot(ucr['fuel_type'])
plt.show()
```

- SCATTER PLOT

```python
# Scattter plot of column 'transmission' and 'torque(Nm)'
sns.scatterplot(data=ucr,x="transmission",y="torque(Nm)")
plt.show
```

- PAIR PLOT

```python
# pair plot 
subset_ucr = ucr.sample(n=100, random_state=14)
sns.pairplot(subset_ucr)
plt.show()
```

- BOX PLOT


```python
# Box plot of column 
sns.boxplot(x=ucr['insurance_validity'], y=ucr['torque(Nm)'], data=ucr)
plt.show()
```

- COUNT PLOT

```python
# Count plot
sns.countplot(x=ucr['engine(cc)'], data=ucr)
plt.show()
```
